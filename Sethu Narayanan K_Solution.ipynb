```python
def playRPSWithComputer():

    print("Welcome to play Rock-Paper-Scissors with computer!")

    # how many rounds player wants to play
    no_of_rounds = int(input("\nNumber of rounds u want to play = "))

    # list contains Rock-Paper-Scissors
    signs = ['Rock','Paper','Scissors']

    # initializing scores to zero
    user_score = 0
    comp_score = 0
    draw_score = 0

    # loop for the number of rounds
    while no_of_rounds>0:
        print("\nRound ",no_of_rounds)
    
        # user playing
        user = input("\nuser: ")
    
        # game is on only if 
        if user == 'Rock' or user == 'Paper' or user == 'Scissors':
        
            # randomizing the computer signs
            comp = random.choice(signs)
            print("\ncomp:",comp)
        
        
            if user == comp:
                draw_score += 1 # Draw
            else:
                if user == 'Rock' and comp == 'Paper':
                    comp_score += 1 # Comp wins
                elif user == 'Rock' and comp == 'Scissors':
                    user_score += 1 # User wins
                elif user == 'Paper' and comp == 'Scissors':
                    comp_score += 1 # Comp wins
                elif user == 'Scissors' and comp == 'Rock':
                    comp_score += 1 # Comp wins
                elif user == 'Scissors' and comp == 'Paper':
                    user_score += 1 # User wins
                else:
                    user_score += 1 # User wins
    
        else:
            print('invalid')
        no_of_rounds -= 1

    # printing the scores
    print("\nuser_score = ",user_score)
    print("comp_score = ",comp_score)
    print("draw_score = ",draw_score)

    # deciding the winner
    if user_score >  comp_score:
        print("\nUser wins")
    elif user_score <  comp_score:
        print("\nComp wins")
    else:
        print("\nDraw")
    
    print("\nThanks for playing.")
```

```python

def choose_random_word():
    words = ["encyclopedia", "anthropology", "stature", "pseudonym", "cryogenic"]
    return random.choice(words)

def display_word(word, guessed_letters):
    display = ""
    for letter in word:
        if letter in guessed_letters:
            display += letter
        else:
            display += "_"
    return display

def word_guessing_game():
    print("Welcome to the Word Guessing Game!")
    
    # Choose a random word
    secret_word = choose_random_word()
    
    # Initialize variables
    guessed_letters = []
    attempts = 5  

    
    while attempts > 0:
        print("\nCurrent word:", display_word(secret_word, guessed_letters))
        
        print("Guessed letters:", guessed_letters)
        
        # Get user input for a letter guess
        guess = input("Guess a letter: ").lower()
        
        # Check if the letter has already been guessed
        if guess in guessed_letters:
            print("You already guessed that letter. Try again.")
            continue
        
        # Add the guessed letter to the list
        guessed_letters.append(guess)
        
        # Check if the guessed letter is in the word
        if guess not in secret_word:
            attempts -= 1
            print("Incorrect! Attempts left:", attempts)
        
        # Check if the player has guessed all letters
        if guessed_letters == secret_word:
            print("Congratulations! You guessed the word:", secret_word)
    
    # If the player runs out of attempts
    if attempts == 0:
        print("Sorry, you ran out of attempts.\n The word was:", secret_word)

```
